## Linear Regression

```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Tips.csv')
df.head()
```

```python
#iporting the reggression library
from sklearn.linear_model import LinearRegression

#Create a Linear regresssion pbject/model using the existing data 
lmodel = LinearRegression()
lmodel
```

```python
#Extracting the features to be used in the prediction
x = df[['total_bill']]
y = df[['tip']]

#Fit the existing data into the regression object/model
lmodel.fit(x,y)
```

```python
#Predict values (tips) using the model
Y = lmodel.predict(x)
Y[0:10]
```

```python
#To know the value value of the intercept
lmodel.intercept_
```

```python
#To know the value of the coefficient
lmodel.coef_
```
